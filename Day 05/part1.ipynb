```python
array = []

with open("input05.txt", 'r') as file:
    array = file.readlines()
```

```python
stacksLength = 0
while array[stacksLength][1]!='1':
    stacksLength+=1
```

```python
xDim = (len(array[stacksLength-1])) // 4

stacks = []
for i in range(xDim):
    stacks.append([])
print(stacks)
```

```python
for i in range(stacksLength-1, -1, -1):
    for j in range(xDim):
        char = array[i][j*4+1]
        if char != ' ':
            stacks[j].append(char)

print(stacks)
```

```python
for command in array[stacksLength+2::]:
    parts = command.split()
    print(parts)
    moveCount = int(parts[1])
    stack1 = int(parts[3])
    stack2 = int(parts[5])
    for i in range(moveCount):
        stacks[stack2-1].append(stacks[stack1-1].pop())
        #stacks[stack1-1].pop()
```

```python
for stack in stacks:
    print(stack[-1])
```
