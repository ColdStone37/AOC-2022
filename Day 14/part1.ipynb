```python
array = []

with open("input14.txt", 'r') as file:
    array = file.readlines()
    
array = [[[int(c) for c in b.split(",")] for b in a.split("->")] for a in array]
print(array)
```

```python
minX = 1000000000
maxX = -1000000000
minY = 1000000000
maxY = -1000000000

for list in array:
    for vector in list:
        if vector[0] < minX:
            minX = vector[0]
        if vector[0] > maxX:
            maxX = vector[0]
        if vector[1] < minY:
            minY = vector[1]
        if vector[1] > maxY:
            maxY = vector[1]
print(minX,maxX,minY,maxY)
```

```python
Map = []
for y in range(maxY+1):
    row = []
    for x in range(maxX-minX+1):
        row.append(".")
    Map.append(row)
```

```python
for list in array:
    for index in range(len(list)-1):
        vector1 = list[index]
        vector2 = list[index+1]
        if vector1[0] == vector2[0]:
            for y in range(min(vector1[1],vector2[1]),1+max(vector1[1],vector2[1])):
                Map[y][vector1[0]-minX] = "#"
        elif vector1[1] == vector2[1]:
            for x in range(min(vector1[0],vector2[0]),1+max(vector1[0],vector2[0])):
                Map[vector1[1]][x-minX] = "#"
        else:
            print("Help")
for row in Map:
    print(row)
```

```python
def simulateSand(x, y, Map):
    if y >= len(Map) - 1:
        return True, [-1,-1]
    if Map[y+1][x] == ".":
        return False, [x,y+1]
    if x <= 0:
        return True, [-1,-1]
    if Map[y+1][x-1] == ".":
        return False, [x-1,y+1]
    if x >= len(Map[y])-1:
        return True, [-1,-1]
    if Map[y+1][x+1] == ".":
        return False, [x+1,y+1]
    return False, [x,y]
```

```python
finished = False
count = 0
while not finished:
    pos = [-1,-1]
    newPos = [500-minX,0]
    
    while not (pos == newPos or finished):
        pos = newPos
        finished, newPos = simulateSand(pos[0],pos[1],Map)
    Map[pos[1]][pos[0]] = "O"
    count += 1
    
for row in Map:
    [print(elem,end="") for elem in row]
    print()
    
print(count-1)
```
