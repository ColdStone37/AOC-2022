```python
array = []

with open("input.txt", 'r') as file:
    array = file.readlines()
```

```python
dir = []
steps = []
for line in array:
    split = line.split(" ")
    dir.append(split[0])
    steps.append(int(split[1]))
print(dir, steps)
```

```python
def constrain(val, min_val, max_val):
    return min(max_val, max(min_val, val))
```

```python
tail = []
for i in range(10):
    tail.append([0,0])

visitedPositions = [[0,0]]

for i, Dir in enumerate(dir):
    moveVector = [0, 0]
    if Dir == "R":
        moveVector[0] = 1
    elif Dir == "L":
        moveVector[0] = -1
    elif Dir == "U":
        moveVector[1] = -1
    elif Dir == "D":
        moveVector[1] = 1
    
    for j in range(steps[i]):
        tail[0][0] += moveVector[0]
        tail[0][1] += moveVector[1]
        
        for k in range(1,10):
            if abs(tail[k-1][0] - tail[k][0]) >= 2 or abs(tail[k-1][1] - tail[k][1]) >= 2:
                tail[k][0] = constrain(tail[k-1][0], tail[k][0]-1, tail[k][0]+1)
                tail[k][1] = constrain(tail[k-1][1], tail[k][1]-1, tail[k][1]+1)
        if not tail[9] in visitedPositions:
            visitedPositions.append([tail[9][0],tail[9][1]])

print(len(visitedPositions))
```
