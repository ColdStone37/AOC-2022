```python
array = []

with open("input09.txt", 'r') as file:
    array = file.readlines()
```

```python
dir = []
steps = []
for line in array:
    split = line.split(" ")
    dir.append(split[0])
    steps.append(int(split[1]))
print(dir, steps)
```

```python
def constrain(val, min_val, max_val):
    return min(max_val, max(min_val, val))
```

```python
head = [0, 0]
tail = [0, 0]

visitedPositions = [[0,0]]

for i, Dir in enumerate(dir):
    moveVector = [0, 0]
    if Dir == "R":
        moveVector[0] = 1
    elif Dir == "L":
        moveVector[0] = -1
    elif Dir == "U":
        moveVector[1] = -1
    elif Dir == "D":
        moveVector[1] = 1
    
    for j in range(steps[i]):
        head[0] += moveVector[0]
        head[1] += moveVector[1]
        
        if abs(head[0] - tail[0]) >= 2 or abs(head[1] - tail[1]) >= 2:
            tail[0] = constrain(head[0], tail[0]-1, tail[0]+1)
            tail[1] = constrain(head[1], tail[1]-1, tail[1]+1)
            if not tail in visitedPositions:
                visitedPositions.append([tail[0],tail[1]])
        

print(len(visitedPositions))
```
